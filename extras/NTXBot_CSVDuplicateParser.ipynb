```python
##########################################
# This script extract the Slack users
# who already subscribed to the newsletter
# with the same email address...
# There will be some that will get the 
# notification because they used 2 diff
# emails...
# 
# For long lists, check out Version 2.
#
# --> Can't send message to 1000 at once.
##########################################
import csv
from operator import itemgetter
from collections import Counter

# 1. Load CSV files (mailing list & slack members)
csvMailingListMembers = open('mailinglist.csv')  # Keep only Slack ID + email.
csvSlackMembers = open('slack.csv') # Keep only email + name

subscribbedEmails = {}
subscribbedMembers = {}

# 2. Display them (visual validation)
for row in csv.reader(csvMailingListMembers):
    print(row)
    # Add emails as Key in subscribbedEmails
    key = row[0]
    subscribbedEmails[key] = row

print(' ')
print(' ')
print('========================================')
print(' ')
print(' ')

count = 0
for row in csv.reader(csvSlackMembers):
    print(row)
    email = row[1]
    if email in subscribbedEmails:
        count = count + 1
        subscribbedMembers[count] = row[0]
    
print(' ')
print(' ')
print('========================================')
print(' ')
print(' ')

for key in subscribbedMembers:
    print(subscribbedMembers[key])

# Writes final CSV, all that is left is to copy it on the server.
with open('allbut.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in subscribbedMembers.items():
        writer.writerow([key, value])
```
